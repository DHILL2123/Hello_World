```python
def clean_titanic(df_titanic):
    '''
    Clean titanic will take in a single pandas dataframe
    and will proceed to drop redundant couns 
    and nonuseful information 
    in addition to addressing null values
    and encoding categorical variables
    '''
    
#impute average age and most common embark_town:
    df_titanic['age'] = df_titanic['age'].fillna(df_titanic.age.mean())
    df_titanic['embark_town'] = df_titanic['embark_town'].fillna('Southhampton')
    #encode categorical values
    df_titanic = pd.concat(
    [df_titanic, pd.get_dummies(df_titanic[['sex','embark_town']], drop_first=True)], axis=1)
    
    return df_titanic
```

```python
def split_titanic_data(df_titanic):
    '''
     split titanic data will split data based on 
    the values present in a cleaned version of titanic
    that is from clean_titanic
    '''
    
    train_val, test = train_test_split(df_titanic, train_size=0.8, random_state=1349, stratify=df_titanic['survived'])
    
    train, validate = train_test_split(train_val, train_size=0.7, random_state=1349, stratify=train_val['survived'])
    
    return train, validate, test
```

```python
def getDuplicateColumns(df_telcochurn):
 
    # Create an empty set
    duplicateColumnNames = set()
     
    # Iterate through all the columns
    # of dataframe
    for x in range(df_telcochurn.shape[1]):
         
        # Take column at xth index.
        col = df_telcochurn.iloc[:, x]
         
        # Iterate through all the columns in
        # DataFrame from (x + 1)th index to
        # last index
        for y in range(x + 1, df_telcochurn.shape[1]):
             
            # Take column at yth index.
            otherCol = df_telcochurn.iloc[:, y]
             
            # Check if two columns at x & y
            # index are equal or not,
            # if equal then adding
            # to the set
            if col.equals(otherCol):
                duplicateColumnNames.add(df_telcochurn.columns.values[y])
                 
    # Return list of unique column names
    # whose contents are duplicates.
    return list(duplicateColumnNames)
```

```python
def train_validate_test_split(iris, target='species', seed=123):
    '''
    This function takes in a dataframe, the name of the target variable
    (for stratification purposes), and an integer for a setting a seed
    and splits the data into train, validate and test. 
    Test is 20% of the original dataset, validate is .30*.80= 24% of the 
    original dataset, and train is .70*.80= 56% of the original dataset. 
    The function returns, in this order, train, validate and test dataframes. 
    '''
    train_validate, test = train_test_split(iris, test_size=0.2, 
                                            random_state=seed, 
                                            stratify=iris[target])
    train, validate = train_test_split(train_validate, test_size=0.3, 
                                       random_state=seed,
                                       stratify=train_validate[target])
    return train, validate, test
```

For virginica & versicolor: Compare the mean petal_width using the Mann-Whitney test (scipy.stats.mannwhitneyu) to see if there is a significant difference between the two groups. Do the same for the other measurement types.

```python
train[train['species']== 'virginica'].petal_width.mean()
train[train['species']== 'versicolor'].petal_width.mean()
```

```python
group1 = train[train['species']== 'virginica'].petal_width.mean()
group2 = train[train['species']== 'versicolor'].petal_width.mean()

stats.mannwhitneyu(group1, group2, alternative='two-sided')
```

In this example, the Mann-Whitney U Test uses the following null and alternative hypotheses:

H0: The petal_width avg is equal between the two groups

HA: The petal_width avg is not equal between the two groups

Since the p-value (1.0) is not less than 0.05, we fail to reject the null hypothesis.

This means we do not have sufficient evidence to say that the true mean petal_width is different between the two groups.
